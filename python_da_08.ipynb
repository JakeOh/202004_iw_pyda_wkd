```
import numpy as np
import pandas as pd
```

```
# 마운트된 Google Drive에서 파일 열기
# 파일 타입에 따라서 pd.read_xxx() 함수 이용
# CSV(Comma-Separated Values): 값들이 쉼표(,)로 구분된 텍스트 파일
#   공백(' '), 탭('\t') 등으로 값들이 구분되어 있는 경우도 있음.
gapminder = pd.read_csv('drive/My Drive/Colab Notebooks/python_da/gapminder.tsv',
                        sep='\t')
gapminder.head()
# DataFrame.head(): 데이터 프레임에서 처음 일부분(기본값 5개)를 출력 
```

```
# DataFrame.tail(): 데이터 프레임에서 마지막 일부분을 출력
gapminder.tail(n=10)
```

```
print(gapminder.shape)
print(gapminder.columns)
print(gapminder.index)
```

```
gapminder[gapminder['country'] == 'Korea, Rep.']
```

```
# Korea, Rep. 중에서 1990년 이상의 데이터만 출력
subset = gapminder[(gapminder['country'] == 'Korea, Rep.') &
                   (gapminder['year'] >= 1990)]
print(subset)
print(subset[['year', 'pop']])
```

```
# Asia 대륙에서 1인당GDP가 20,000 이상인 데이터
gapminder[(gapminder['continent'] == 'Asia') &
          (gapminder['gdpPercap'] >= 20_000)]
```
